```python
import numpy as np
import pandas as pd
# pip install seaborn 
import seaborn as sns
import matplotlib.pyplot as plt

dtype = {'DayOfWeek': np.uint8, 'DayofMonth': np.uint8, 'Month': np.uint8 , 'Cancelled': np.uint8, 
         'Year': np.uint16, 'FlightNum': np.uint16 , 'Distance': np.uint16, 
         'UniqueCarrier': str, 'CancellationCode': str, 'Origin': str, 'Dest': str,
         'ArrDelay': np.float16, 'DepDelay': np.float16, 'CarrierDelay': np.float16,
         'WeatherDelay': np.float16, 'NASDelay': np.float16, 'SecurityDelay': np.float16,
         'LateAircraftDelay': np.float16, 'DepTime': np.float16}

flights_df = pd.read_csv("2008.csv", usecols = dtype.keys(), dtype = dtype)



def timespace_category(time):
    if 0 <= time <= 59:
        return '0'
    elif 100 <= time <= 159:
        return '1'
    elif 200 <= time <= 259:
        return '2'
    elif 300 <= time <= 359:
        return '3'
    elif 400 <= time <= 459:
        return '4'
    elif 500 <= time <= 559:
        return '5'
    elif 600 <= time <= 659:
        return '6'
    elif 700 <= time <= 759:
        return '7'
    elif 800 <= time <= 859:
        return '8'
    elif 900 <= time <= 959:
        return '9'
    elif 1000 <= time <= 1059:
        return '10'
    elif 1100 <= time <= 1159:
        return '11'
    elif 1200 <= time <= 1259:
        return '12'
    elif 1300 <= time <= 1359:
        return '13'
    elif 1400 <= time <= 1459:
        return '14'
    elif 1500 <= time <= 1559:
        return '15'
    elif 1600 <= time <= 1659:
        return '16'
    elif 1700 <= time <= 1759:
        return '17'
    elif 1800 <= time <= 1859:
        return '18'
    elif 1900 <= time <= 1959:
        return '19'
    elif 2000 <= time <= 2059:
        return '20'
    elif 2100 <= time <= 2159:
        return '21'
    elif 2200 <= time <= 2259:
        return '22'
    elif 2300 <= time <= 2359:
        return '23'
    return ''


#flights_df.head(30)


# map() applies age_category() function to every value in flights_df.DepTime
flights_df['timespace_category'] = flights_df.DepTime.map(timespace_category)


flights_df.groupby('timespace_category').size().plot(kind='bar');
```
